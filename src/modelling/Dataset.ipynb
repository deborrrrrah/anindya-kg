```
import pandas as pd 
import numpy as np
import json
from seqeval.metrics import classification_report
```

```
def readJSON(filename) :
  with open(filename, 'r', encoding="utf8") as f:
    obj = json.load(f)
  return obj

def writeJSON(obj, filename) :
  with open(filename, 'w', encoding="utf8") as outfile:
    json.dump(obj, outfile)
  print ("Successfully write JSON obj to", filename)

def print_support(version, files) :
  for filename in files :
    print (filename)
    data = readJSON(ROOT_PATH + "dataset/version ["+ str(version) + "]/" + filename)
    labels = [[object['label'] for object in objects] for objects in data["tokens_labels"]]
    print (classification_report(labels, labels))
```

```
files = ["scenario-1-test.json", "scenario-1-train.json", "scenario-2-test.json"]
```

```
print_support(1, files)
```
